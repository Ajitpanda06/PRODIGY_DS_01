```python
# For demonstration, let's create a sample DataFrame
data = pd.DataFrame({
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Male', 'Female', 'Female'],
    'Count': [25, 30, 35, 40, 45, 50, 55, 60, 65, 70]
})
```

```python
# Group the data by 'Gender' and calculate the sum of counts
gender_counts = data.groupby('Gender')['Count'].sum()
```

```python
# Plot the bar chart
plt.figure(figsize=(8, 6))
gender_counts.plot(kind='bar', color=['blue', 'pink'])
plt.title('Distribution of Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
# Sample data (ages)
ages = [25, 30, 35, 40, 45, 50, 55, 60, 65, 70]

# Plot histogram
plt.hist(ages, bins=5, color='skyblue', edgecolor='black')
plt.title('Distribution of Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Sample data (ages)
ages = [25, 30, 35, 40, 45, 50, 55, 60, 65, 70]

# Plot box plot
sns.boxplot(x=ages, color='skyblue')
plt.title('Box Plot of Ages')
plt.xlabel('Age')
plt.show()
```

```python
# Sample data (genders)
genders = ['Male', 'Female']
counts = [60, 40]

# Plot pie chart
plt.pie(counts, labels=genders, colors=['blue', 'pink'], autopct='%1.1f%%')
plt.title('Distribution of Genders')
plt.show()
```

```python
# Sample data (heights and weights)
heights = [160, 165, 170, 175, 180]
weights = [60, 65, 70, 75, 80]

# Plot scatter plot
plt.scatter(heights, weights, color='skyblue')
plt.title('Scatter Plot of Heights vs Weights')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.show()
```

```python
# Sample correlation matrix
corr_matrix = np.array([[1.0, 0.5, -0.3],
                        [0.5, 1.0, 0.7],
                        [-0.3, 0.7, 1.0]])

# Plot heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', square=True)
plt.title('Correlation Heatmap')
plt.show()
```

```python

# Define the years from 1960 to 2022
years = list(range(1960, 2023))

# Sample population data (replace it with your actual population data)
# For demonstration, let's assume a linearly increasing population
population = [100 + i * 5 for i in range(len(years))]  # Replace this with your actual population data

# Plot the line plot
plt.plot(years, population, marker='o', color='skyblue')
plt.title('Population Trend')
plt.xlabel('Year')
plt.ylabel('Population (millions)')
plt.xticks(range(1960, 2023, 10))  # Show every 10 years on the x-axis
plt.grid(True)
plt.show()
```

```python
# Sample data (synthetic dataframe)
data = pd.DataFrame({
    'A': [1, 2, 3, 4, 5],
    'B': [5, 4, 3, 2, 1],
    'C': [3, 3, 3, 3, 3]
})

# Plot pair plot
sns.pairplot(data)
plt.title('Pairwise Scatter Plot')
plt.show()
```

```python
# Sample data (categories and counts)
categories = ['Category 1', 'Category 2']
counts_male = [30, 20]
counts_female = [20, 30]

# Plot stacked bar chart
plt.bar(categories, counts_male, color='blue', label='Male')
plt.bar(categories, counts_female, bottom=counts_male, color='pink', label='Female')
plt.title('Distribution of Categories by Gender')
plt.xlabel('Category')
plt.ylabel('Count')
plt.legend()
plt.show()
```

```python
population = [100 + (year - 1960) * 5 for year in years]  # Replace this with your actual population data

# Plot the area chart
plt.fill_between(years, population, color='skyblue', alpha=0.4)
plt.plot(years, population, color='skyblue', alpha=0.6)
plt.title('Population Trend')
plt.xlabel('Year')
plt.ylabel('Population (millions)')
plt.xticks(range(1960, 2023, 10))  # Show every 10 years on the x-axis
plt.grid(True)
plt.show()
```
